```python
import pandas as pd
import matplotlib.pyplot as plt

file = "TEK0008.CSV"

# Lees CSV in en skip de eerste 20 regels metadata (Tektronix standaard)
df = pd.read_csv(file, skiprows=20, header=None)

# Haal tijd en signaal uit de juiste kolommen
x = pd.to_numeric(df[3], errors="coerce")
y = pd.to_numeric(df[4], errors="coerce")

# Verwijder NaN-rijen
mask = (~x.isna()) & (~y.isna())
x = x[mask]
y = y[mask]

# Plot
plt.figure(figsize=(10,5))
plt.plot(x, y)
plt.title("Rubidium Spectrum")
plt.xlabel("Scan (tijd/frequentie)")
plt.ylabel("Signaal (V)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
