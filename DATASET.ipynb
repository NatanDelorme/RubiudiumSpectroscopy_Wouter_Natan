```python
import pandas as pd
import matplotlib.pyplot as plt

file1 = "PD1_DF_DIP2.CSV"
file2 = "PD1_DB_DIP2.CSV"

# Eerste CSV
df1 = pd.read_csv(file1, skiprows=20, header=None)
x1 = pd.to_numeric(df1[3], errors="coerce")
y1 = pd.to_numeric(df1[4], errors="coerce")
mask1 = (~x1.isna()) & (~y1.isna())
x1 = x1[mask1]
y1 = y1[mask1]

# Tweede CSV
df2 = pd.read_csv(file2, skiprows=20, header=None)
x2 = pd.to_numeric(df2[3], errors="coerce")
y2 = pd.to_numeric(df2[4], errors="coerce")
mask2 = (~x2.isna()) & (~y2.isna())
x2 = x2[mask2]
y2 = y2[mask2]

# Plot beide lijnen in n figuur
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, label="Spectrum 1")
plt.plot(x2, y2, label="Spectrum 2")

plt.title("Rubidium Spectrum")
plt.xlabel("Scan (tijd/frequentie)")
plt.ylabel("Signaal (V)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
