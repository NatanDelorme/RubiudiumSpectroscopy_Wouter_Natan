```python
import pandas as pd
import matplotlib.pyplot as plt

file1 = "PD1_DF_DIP1.CSV"
file2 = "PD1_DB_DIP1.CSV"

# Eerste CSV
df1 = pd.read_csv(file1, skiprows=20, header=None)
x1 = pd.to_numeric(df1[3], errors="coerce")
y1 = pd.to_numeric(df1[4], errors="coerce")
mask1 = (~x1.isna()) & (~y1.isna())
x1 = x1[mask1]
y1 = y1[mask1]

# Tweede CSV
df2 = pd.read_csv(file2, skiprows=20, header=None)
x2 = pd.to_numeric(df2[3], errors="coerce")
y2 = pd.to_numeric(df2[4], errors="coerce")
mask2 = (~x2.isna()) & (~y2.isna())
x2 = x2[mask2]
y2 = y2[mask2]

# Plot beide lijnen in n figuur
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, label="Doppler free spectrum")
plt.plot(x2, y2, label="Doppler broadened spectrum")

plt.title("Rubidium Spectrum PD1 DIP1")
plt.xlabel("Scan (tijd/frequentie)")
plt.ylabel("Signaal (V)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file1 = "PD1_DF_DIP1.CSV"
file2 = "PD1_DB_DIP1.CSV"

# Eerste CSV
df1 = pd.read_csv(file1, skiprows=20, header=None)
x1 = pd.to_numeric(df1[3], errors="coerce")
y1 = pd.to_numeric(df1[4], errors="coerce")
mask1 = (~x1.isna()) & (~y1.isna())
x1 = x1[mask1].reset_index(drop=True)
y1 = y1[mask1].reset_index(drop=True)

# Tweede CSV
df2 = pd.read_csv(file2, skiprows=20, header=None)
x2 = pd.to_numeric(df2[3], errors="coerce")
y2 = pd.to_numeric(df2[4], errors="coerce")
mask2 = (~x2.isna()) & (~y2.isna())
x2 = x2[mask2].reset_index(drop=True)
y2 = y2[mask2].reset_index(drop=True)

# Zorg dat de lengtes gelijk zijn
n = min(len(y1), len(y2))
x = x1[:n]          # neem de eerste n punten van x1
y_diff =  y2[:n] - y1[:n]   # trek de spanningen van elkaar af

# Plot verschil
plt.figure(figsize=(10, 5))
plt.plot(x, y_diff, label="Doppler free - Doppler broadened")

plt.title("Verschil in rubidium spectrum PD1 DIP1")
plt.xlabel("Scan (tijd/frequentie)")
plt.ylabel("Verschil in signaal (V)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
